### CRU


```python

# Configuration des chemins des dossiers contenant les données
path_base_cru = "dataset/onedrive DS/cru_data_selected"
liste_domaines = ['cld', 'pre', 'tmp', 'wet', 'frs', 'pet', 'dtr', 'tmx', 'tmn', 'vap']

fao_countries = [
    'Albania', 'Austria', 'Belarus', 'Belgium', 'Bosnia-Herzegovinia', 'Bulgaria',
    'Croatia', 'Czech_Republic', 'Denmark', 'Estonia', 'Faeroes', 'Finland', 
    'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 
    'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Macedonia', 'Montenegro', 
    'Netherlands', 'Norway', 'Poland', 'Portugal', 'Moldova', 'Romania', 
    'Serbia', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine',
    'United_Kingdom'
]

# Dictionnaire de renommage de 'ANN' 
regles_ann = {
    'wet': 'wet_days_total',      # Fréquence
    'frs': 'frost_days_total',     # Fréquence
    'pre': 'pre_total_mm',        # Cumul    
    'pet': 'pet_total_mm',        # Cumul
    'tmp': 'tmp_mean_c',          # Moyenne
    'cld': 'cld_mean_pct',        # Moyenne
    'vap': 'vap_pressure_mean_hpa',# Moyenne
    'dtr': 'diurnal_range_mean_c', # Moyenne
    'tmx': 'tmp_max_mean_c',      # Moyenne
    'tmn': 'tmp_min_mean_c'       # Moyenne
}

df_cru_global = None

# Boucle principale par dossier de CRU (liste_domaines)
for domaine in liste_domaines:
    print(f"🔄 Traitement du sous-domaine : {domaine}...")
    path_domaine = os.path.join(path_base_cru, domaine)
    fichiers_du_dossier = os.listdir(path_domaine)
    
    # Liste pour stocker temporairement les DataFrames des 39 pays pour ce domaine
    df_domaine_pays = []
    # Boucle dans les fichiers du dossier "domaine"
    for country in fao_countries:
        # Recherche du nom dd pays dans la liste des fichiers
        matches = [f for f in fichiers_du_dossier if country.lower() in f.lower()]
        
        if matches:
            fichier_cible = matches[0]
            chemin_fichier = os.path.join(path_domaine, fichier_cible)
            
            df_res = pd.read_csv(chemin_fichier, sep=';', skiprows=3, na_values='-999.0')
            df_res.columns = df_res.columns.str.strip()
            
            # on ne garde que YEAR et ANN
            df_res = df_res[['YEAR', 'ANN']].copy()
            
            # Renommage de ANN avec le nom unique du domaine 
            nom_feature = regles_ann[domaine]
            df_res = df_res.rename(columns={'YEAR': 'Year', 'ANN': nom_feature})
            
            # création de la colonne pays/area FAO 
            df_res['Area'] = country
            
            df_domaine_pays.append(df_res)
            
    # On rassemble verticalement tous les pays pour ce domaine précis
    df_un_domaine_tous_pays = pd.concat(df_domaine_pays, ignore_index=True)
    
    # On fusionne horizontalement ce domaine avec les précédents sur (Area, Year)
    if df_cru_global is None:
        df_cru_global = df_un_domaine_tous_pays
    else:
        df_cru_global = pd.merge(df_cru_global, df_un_domaine_tous_pays, on=['Area', 'Year'], how='outer')

```

       'Herbicides', 'Insecticides', 'Pesticides (total)', 'Rodenticides'


```python
# Pivotement des données de nutriment pour passer les différents types en colonnes distinctes
fertiliz_pivot = fao_lis_fertiliz_data.pivot_table(
    index=['Area', 'Year'],  # Clé unique par ligne : un pays pour une année donnée
    columns='Item',          # Chaque type d'engrais devient un en-tête de colonne
    values='Value',
    aggfunc='first'
).reset_index()              # On remet Area et Year en colonnes standard au lieu de les laisser en index

# Même opération sur les données PIHFR (pesticides, etc.) pour aligner la structure
pihfr_pivot = fao_lis_pihfr_data.pivot_table(
    index=['Area', 'Year'], 
    columns='Item', 
    values='Value',
    aggfunc='first'
).reset_index()
```

```python
# Merge Final des 4 dataframes/domaines
# la jointure se fait par les années et pays
df_merge_1 = pd.merge(fao_Crops_Livestock_E_Europe, fertiliz_pivot, on=['Area', 'Year'], how='left')

df_merge_2 = pd.merge(df_merge_1, pihfr_pivot, on=['Area', 'Year'], how='left')

df_final_agritech = pd.merge(df_merge_2, df_cru_global, on=['Area', 'Year'], how='left')

# Vérification rapide de la dimension de la matrice 
print(f"Nombre de colonnes (Features dispo pour l'EDA) : {df_final_agritech.shape[1]}")
```
